```python
import pandas as pd
import numpy as np

import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

# A. Build a baseline model 

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python

y = concrete_data['Strength']
X = concrete_data.drop(columns=['Strength'])
print(y)
```

# Let's check how many data points we have.

```python
X.shape
```

# Let's check the dataset for any missing values.

```python
X.isnull().sum()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.3, random_state=3)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape 
```

```python
n_cols = X_train.shape[1] # number of predictorsn_cols
n_cols
```

# Build a Neural Network

```python
# define regression model
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    #model.add(Dense(50, activation='relu'))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
    return model
```

```python
# build the model
print(regression_model())
model = regression_model()
```

```python
# fit the model
list = []
for x in range (50):
    model.fit(X_train, y_train, epochs=50, verbose=1)
    y_pred = model.predict(X_test)
    acc = mean_squared_error(y_test, y_pred.round())
    list.append(acc)
    
```

# Mean and the standard deviation of the mean squared errors

```python
print (list)
```

```python
scaler = StandardScaler()
arr = np.array(list)
ar = arr.reshape(-1, 1)
scaler.fit(ar)
print(scaler.mean_)
```

# B. Normalize the data

```python
norm = (concrete_data - concrete_data.mean()) / concrete_data.std()
norm.head()
```

```python
y1 = norm['Strength']
X1 = norm.drop(columns=['Strength'])
```

```python
X1_train,X1_test,y1_train,y1_test = train_test_split(X1, y1, test_size=0.3, random_state=3)
```

```python
# define regression model
def regression_model2():
    # create model
    model2 = Sequential()
    model2.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    #model2.add(Dense(50, activation='relu'))
    model2.add(Dense(1))
    
    # compile model
    model2.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
    return model2

# build the model
print(regression_model2())
model2 = regression_model2()
```

```python
# fit the model
list2 = []
for x in range (50):
    model2.fit(X1_train, y1_train, epochs=50, verbose=1)
    y1_pred = model2.predict(X1_test)
    acc2 = mean_squared_error(y1_test, y1_pred.round())
    list2.append(acc2)
    
```

```python

scaler2 = StandardScaler()
arr2 = np.array(list2)
ar2 = arr2.reshape(-1, 1)
scaler2.fit(ar2)
print(scaler2.mean_)
```

# D. Increase the number of hidden layers 

```python
def regression_model3():
    # create model
    model2 = Sequential()
    model2.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model2.add(Dense(10, activation='relu'))
    model2.add(Dense(10, activation='relu'))
    model2.add(Dense(1))
    
    # compile model
    model2.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
    return model2

# build the model
print(regression_model3())
model3 = regression_model3()
```

```python
list4 = []
for x in range (50):
    model3.fit(X1_train, y1_train, epochs=100, verbose=1)
    y1_pred = model3.predict(X1_test)
    acc3 = mean_squared_error(y1_test, y1_pred.round())
    list4.append(acc3)
```
